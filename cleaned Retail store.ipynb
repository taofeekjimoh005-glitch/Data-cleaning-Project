```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Fill missing item names with 'Unknown'
df['Item'] = df['Item'].fillna('Unknown')
df.isnull().sum()
```

```python
# Fill missing prices with median price
df['Price Per Unit'] = df['Price Per Unit'].fillna(df['Price Per Unit'].median())
df.isnull().sum()
```

```python
# Fill missing quantity values with median quantity
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df.isnull().sum()
```

```python
# Recalculate total amount spent
df['Total Spent'] = df['Price Per Unit'] * df['Quantity']
df.isnull().sum()
```

```python
# Fill missing discount values with False
df['Discount Applied'] = df['Discount Applied'].fillna('False')
df.isnull().sum()
```

```python
# Convert Transaction Date to datetime
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df.isnull().sum()
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
df.to_csv("Cleaned Retail Store sales.csv", index = False)
```
